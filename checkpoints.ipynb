```python
import kaleido
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
pathing=Path('resources/acs2017_census_tract_data.csv')
census_df= pd.read_csv(pathing, encoding="utf-8")
```

```python
us_census= census_df.drop(census_df[census_df["State"]=="Puerto Rico"].index)
us_census
```

```python
us_df=us_census.drop(columns=["IncomePerCapErr","IncomeErr","IncomePerCap"])
us_df
```

```python
new_us_census=us_census.rename(columns={"TractId":"Tract ID","TotalPop":"Total Population","VotingAgeCitizen":"Voting Age Citizen",
                                   "ChildPoverty":"Child Poverty","OtherTransp":"Other Transportation","WorkAtHome":"Work At Home",
                                   "MeanCommute":"Average Commute","PrivateWork":"Private Work","PublicWork":"Public Work",
                                   "SelfEmployed": "Self-Employed","FamilyWork":"Family Work"})

new_us_census
```

```python
us_df=new_us_census.drop(columns=["IncomePerCapErr","IncomeErr","IncomePerCap"])
us_df
```

```python
income=us_df[["State","County","Total Population","Income"]]
income_county=income.groupby(["State","County"]).sum()
income_county=income_county.reset_index()
income_county
```

```python
finding_age=us_df[["State","County","Total Population","Voting Age Citizen"]]
finding_age
```

```python
find_age=finding_age.groupby("State").sum()
find_age=find_age.reset_index()
find_age
```

```python
find_age['Under 18 Citizen']=find_age['Total Population']-find_age['Voting Age Citizen']

age_population=find_age.rename(columns={"Voting Age Citizen":"Over 18 Citizen"})
age_population["Total Population"] = age_population["Total Population"].map("{:,}".format)
age_population["Over 18 Citizen"] = age_population["Over 18 Citizen"].map("{:,}".format)
age_population["Under 18 Citizen"] = age_population["Under 18 Citizen"].map("{:,}".format)


age_population
```

```python
age_population.to_csv("static/csv_files/age_population.csv",
                  encoding="utf-8", index=False, header=True)
```

```python
population=us_df[["Tract ID","State","County","Total Population","Men","Women","Hispanic","White","Black","Native","Asian","Pacific"]]
population
```

```python
df=us_df.groupby(['State','County']).sum()
df
```

```python
hispanic_deci=population["Hispanic"]/100
white_deci=population["White"]/100
black_deci=population["Black"]/100
native_deci=population["Native"]/100
asian_deci=population["Asian"]/100
pacific_deci=population["Pacific"]/100


tract=population["Tract ID"]
state=population["State"]
county=population["County"]
popu=population["Total Population"]
men=population["Men"]
women=population["Women"]


popul=pd.DataFrame({"State":state,"County":county,"Total Population":popu,"Men":men,"Women":women,
                    "Hispanic":hispanic_deci,"White":white_deci,"Black":black_deci,"Native":native_deci,
                    "Asian":asian_deci,"Pacific":pacific_deci})
popul
```

```python
his_women=popul.Hispanic*popul.Women
hispanic_women=his_women.fillna(0)
new_hispanic_women=hispanic_women.astype(int)

his_men=popul.Men*popul.Hispanic
hispanic_men=his_men.fillna(0)
new_hispanic_men=hispanic_men.astype(int)


hispanic_population=pd.DataFrame({"State":state,"County":county,"Total Population":popu,
                                  "Hispanic Men":new_hispanic_men,"Hispanic Women":new_hispanic_women})

hispanic_pop=hispanic_population.groupby(['State','County']).sum()
hispanic_pop["Hispanic Men"] = hispanic_pop["Hispanic Men"].map("{:,}".format)
hispanic_pop["Hispanic Women"] = hispanic_pop["Hispanic Women"].map("{:,}".format)

hispanic_pop
```

```python
whi_women=popul.Women*popul.White
white_women=whi_women.fillna(0)
new_white_women=white_women.astype(int)

whi_men=popul.Men*popul.White
white_men=whi_men.fillna(0)
new_white_men=white_men.astype(int)


white_population=pd.DataFrame({"State":state,"County":county,"Total Population":popu,
                                  "White Men":new_white_men,"White Women":new_white_women})

white_pop=white_population.groupby(['State','County']).sum()
white_pop["White Men"] = white_pop["White Men"].map("{:,}".format)
white_pop["White Women"] = white_pop["White Women"].map("{:,}".format)


white_pop
```

```python
bla_women=popul.Women*popul.Black
black_women=bla_women.fillna(0)
new_black_women=black_women.astype(int)

bla_men=popul.Men*popul.Black
black_men=bla_men.fillna(0)
new_black_men=black_men.astype(int)


black_population=pd.DataFrame({"State":state,"County":county,"Total Population":popu,
                                  "Black Men":new_black_men,"Black Women":new_black_women})

black_pop=black_population.groupby(['State','County']).sum()
black_pop["Black Men"] = black_pop["Black Men"].map("{:,}".format)
black_pop["Black Women"] = black_pop["Black Women"].map("{:,}".format)

black_pop
```

```python
nat_women=popul.Women*popul.Native
native_women=nat_women.fillna(0)
new_native_women=native_women.astype(int)

nat_men=popul.Men*popul.Native
native_men=nat_men.fillna(0)
new_native_men=native_men.astype(int)


native_population=pd.DataFrame({"State":state,"County":county,"Total Population":popu,
                                  "Native Men":new_native_men,"Native Women":new_native_women})

native_pop=native_population.groupby(['State','County']).sum()
native_pop["Native Men"] = native_pop["Native Men"].map("{:,}".format)
native_pop["Native Women"] = native_pop["Native Women"].map("{:,}".format)

native_pop
```

```python
asi_women=popul.Women*popul.Asian
asian_women=asi_women.fillna(0)
new_asian_women=asian_women.astype(int)

asi_men=popul.Men*popul.Asian
asian_men=asi_men.fillna(0)
new_asian_men=asian_men.astype(int)

asian_population=pd.DataFrame({"State":state,"County":county,"Total Population":popu,
                                  "Asian Men":new_asian_men,"Asian Women":new_asian_women})

asian_pop=asian_population.groupby(['State','County']).sum()
asian_pop["Asian Men"] = asian_pop["Asian Men"].map("{:,}".format)
asian_pop["Asian Women"] = asian_pop["Asian Women"].map("{:,}".format)

asian_pop
```

```python
paci_women=popul.Women*popul.Pacific
pacific_women=paci_women.fillna(0)
new_pacific_women=pacific_women.astype(int)

paci_men=popul.Men*popul.Pacific
pacific_men=paci_men.fillna(0)
new_pacific_men=pacific_men.astype(int)


pacific_population=pd.DataFrame({"State":state,"County":county,"Total Population":popu,
                                  "Pacific Men":new_pacific_men,"Pacific Women":new_pacific_women})

pacific_pop=pacific_population.groupby(['State','County']).sum()
pacific_pop["Pacific Men"] = pacific_pop["Pacific Men"].map("{:,}".format)
pacific_pop["Pacific Women"] = pacific_pop["Pacific Women"].map("{:,}".format)

pacific_pop
```

```python
population_df=pd.DataFrame({"States":state,"County":county,"Total Population":popu,
                            "Hispanic Men":new_hispanic_men,"Hispanic Women":new_hispanic_women,
                            "White Men":new_white_men,"White Women":new_white_women,"Black Men":new_black_men,
                            "Black Women":new_black_women, "Native Men":new_native_men,
                            "Native Women":new_native_women,"Asian Men":new_asian_men,
                            "Asian Women":new_asian_women,"Pacific Men":new_pacific_men,
                            "Pacific Women":new_pacific_women})



population_df
```

```python
population_df['Others']=popu-(new_hispanic_men+new_hispanic_women+new_white_men+new_white_women+new_black_men+new_black_women
                              +new_native_men+new_native_women+new_asian_men+new_asian_women+new_pacific_men+new_pacific_women)
population_df
```

```python
county_pop_df=population_df.groupby(["States","County"]).sum()

county_pop_df["Total Population"] = county_pop_df["Total Population"].map("{:,}".format)
county_pop_df["Hispanic Men"] = county_pop_df["Hispanic Men"].map("{:,}".format)
county_pop_df["Hispanic Women"] = county_pop_df["Hispanic Women"].map("{:,}".format)
county_pop_df["White Men"] = county_pop_df["White Men"].map("{:,}".format)
county_pop_df["White Women"] = county_pop_df["White Women"].map("{:,}".format)
county_pop_df["Black Men"] = county_pop_df["Black Men"].map("{:,}".format)
county_pop_df["Black Women"] = county_pop_df["Black Women"].map("{:,}".format)
county_pop_df["Native Men"] = county_pop_df["Native Men"].map("{:,}".format)
county_pop_df["Native Women"] = county_pop_df["Native Women"].map("{:,}".format)
county_pop_df["Asian Men"] = county_pop_df["Asian Men"].map("{:,}".format)
county_pop_df["Asian Women"] = county_pop_df["Asian Women"].map("{:,}".format)
county_pop_df["Pacific Men"] = county_pop_df["Pacific Men"].map("{:,}".format)
county_pop_df["Pacific Women"] = county_pop_df["Pacific Women"].map("{:,}".format)
county_pop_df["Others"] = county_pop_df["Others"].map("{:,}".format)

county_pop_df=county_pop_df.reset_index()

county_pop_df
```

```python
county_pop_df.to_csv("static/csv_files/county_population.csv",
                  encoding="utf-8", index=False, header=True)
```

```python
state_pop_df=population_df.groupby("States").sum()

state_pop_df["Total Population"] = state_pop_df["Total Population"].map("{:,}".format)
state_pop_df["Hispanic Men"] = state_pop_df["Hispanic Men"].map("{:,}".format)
state_pop_df["Hispanic Women"] = state_pop_df["Hispanic Women"].map("{:,}".format)
state_pop_df["White Men"] = state_pop_df["White Men"].map("{:,}".format)
state_pop_df["White Women"] = state_pop_df["White Women"].map("{:,}".format)
state_pop_df["Black Men"] = state_pop_df["Black Men"].map("{:,}".format)
state_pop_df["Black Women"] = state_pop_df["Black Women"].map("{:,}".format)
state_pop_df["Native Men"] = state_pop_df["Native Men"].map("{:,}".format)
state_pop_df["Native Women"] = state_pop_df["Native Women"].map("{:,}".format)
state_pop_df["Asian Men"] = state_pop_df["Asian Men"].map("{:,}".format)
state_pop_df["Asian Women"] = state_pop_df["Asian Women"].map("{:,}".format)
state_pop_df["Pacific Men"] = state_pop_df["Pacific Men"].map("{:,}".format)
state_pop_df["Pacific Women"] = state_pop_df["Pacific Women"].map("{:,}".format)
state_pop_df["Others"] = state_pop_df["Others"].map("{:,}".format)

state_pop_df=state_pop_df.reset_index()

state_pop_df
```

```python
state_pop_df.to_csv("static/csv_files/state_population.csv",
                  encoding="utf-8", index=False, header=True)
```
